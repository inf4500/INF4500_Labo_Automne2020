# Figures et python

La librairie pour produire des figures en python est appelé [*matplotlib*](https://matplotlib.org/). La synthax est un peu étrange mais c'est un outil incroyablement puissant. Un *wrapper* plus avancé que l'on peut rajouter par dessus matplotlib est [*seaborn*](https://seaborn.pydata.org/).

Le principe général est de 

1. Définir la figures / axes etc...
2. "Dessiner" ou sauver la figure
3. Effacer le tout avant d'en faire une autre (sinon elle vont se superposer).

Tout se fait à travers l'objet `plt` de matplotlib, comme suit:

```python
from matplotlib import pyplot as plt
```

# Un graph

```python
#Commençons par générer un ensemble de coordonées X et Y de même longueur
#En matplotlib, les coordonées doivent être séparées une liste à la fois.
#Pour représenter un point (1, 2) il faut faire deux lists
liste_x = [1]
liste_y = [2]
plt.plot(liste_x, liste_y, marker='x') #Faire le dessin, on veut que le point soit un "x"
plt.show()#L'afficher utilisert "plt.savefig("fig.pdf")" pour sauver la figure!!!
plt.clf() #Effacer de la mémoire

```

```python
#Donc une liste de points, nous devons séparer les X et Y.
liste_x = list(range(10)) #0, 1, ... 9
liste_y = [z**2 for z in liste_x]
plt.plot(liste_x, liste_y) #remarquer, sans l'argument "marker"
plt.show()
plt.clf()
```

```python
#On peut rajoute des noms sur les axes et titre
#Donc une liste de points, nous devons séparer les X et Y.
liste_x = list(range(10)) #0, 1, ... 9
liste_y = [z**2 for z in liste_x]
plt.plot(liste_x, liste_y) #remarquer, sans l'argument "marker"
plt.xlabel("Les valeurs de X")
plt.ylabel("Le carré de X")
plt.title("Un bon titre")
plt.show()
plt.clf()
```

# Barrre d'erreur

```python
#On peut rajoute une troisieme liste de valeur qui est l'erreur! https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.errorbar.html
#Donc une liste de points, nous devons séparer les X et Y.
liste_x = list(range(10)) #0, 1, ... 9
liste_y = [z**2 for z in liste_x]
liste_err = [0.05 * z for z in liste_y]

plt.errorbar(liste_x, liste_y, liste_err) #remarquer, sans l'argument "marker"
plt.xlabel("Les valeurs de X")
plt.ylabel("Le carré de X")
plt.title("en plus on a 5% d'erreur!")
plt.show()
plt.clf()
```

# Bar plot

```python
# Un autre truc utile est de pouvoir faire des "bar plots"
#Utilisons deux nouvelles librairies standard de python sympathique
from random import choice
#On veut un ensemble de 1000 elements au hasard entre a,b,c,d,e
data = [choice('abcde') for _ in range(1000)]

#On veut savoir combien de chacun on a! On pourrait compter a la main... c'est long. 
#Vive la librairie collection!
from collections import Counter
data_count = Counter(data)
print(data_count)
```

```python
# A PARTIR DE PYTHON 3.6 L'ORDRE EST TOUJOURS LE MEME POUR CLEFS ET VALEURS!
valeurs = data_count.values()
clefs = data_count.keys()
#Les paires correspondent bien!
print(valeurs)
print(clefs)
```

```python
#Maintenant pour faire un "bar plot", c'est tout simple
plt.bar(clefs, valeurs)
plt.show()
plt.clf()
```
